```python
import cv2
import time
vidcap = cv2.VideoCapture("a.mp4")
#read the frame1
ret, frame1= vidcap.read() #reads the next frame
vehile = 0
while vidcap.isOpened():
    #read the frame 2
    ret, frame2 = vidcap.read()  #read the next frame
    frame = frame2.copy() #duplicate copy for visualization
    #check if end of the frame has been reached
    if not ret:
        break
    #extract the foreground mask
    fgmask = cv2.absdiff(frame1, frame2)
    fgmask = cv2.cvtColor(fgmask, cv2.COLOR_BGR2GRAY)
    
    #apply the threshold for increasingwhiteforeground
    _, thresh = cv2.threshold(fgmask, 100, 255, cv2.THRESH_BINARY)
    #assign frame2 to frame 1 to continue the iteration untill alll frames are read
    frame1 = frame2
    
    #draw the reference traffic lines
    cv2.line(frame, (3,450), (1500,450), (0, 0, 255), 2) #red line
    cv2.line(frame, (3,440), (1500,440), (0, 255, 0), 1) #green line
   
    #extract the contours
    conts, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    
    for c in conts:
        if cv2.contourArea(c) < 100:
            continue
        x, y, w, h = cv2.boundingRect(c)
        
        #draw the bounding rectangle for all contours
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 2)
        xMid = int((x + (x+w))/2)
        yMid = int((y + (y+h))/2)
        cv2.circle(frame, (xMid,yMid),5,(0,0,255),5)
        if yMid > 440 and yMid < 460:
            vehile += 1
        if conts == 1:
            cv2.line(frame, (3,450), (1500,450), (0, 0, 255), 2) 
        else:
            cv2.line(frame, (3,440), (1500,440), (0, 255, 0), 1)                              
    #show the thresh and original vedio
    cv2.imshow('Foreground mask', fgmask)
    cv2.putText(frame, 'Total vehiles : {}'.format(vehile), (450, 50), cv2.FONT_HERSHEY_PLAIN, 2, (255,255,255), 2)
    cv2.imshow('Original video', frame)
    #wait for any key to be pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    #release video capture
cv2.destroyAllWindows()
vidcap.release()
```
